This project models how different levels of Universal Basic Income (UBI) affect aggregate labor supply. It uses Python to simulate the effect based on varying labor-supply elasticities and visualizes the impact using data analysis and plots.

```python
# PARAMETERS
base_income = 50000  # Annual income before UBI
ubi_levels = [0, 6000, 12000, 18000]  # UBI levels to test ($/year)
elasticities = [0.05, 0.10, 0.20]  # Labor supply elasticities (low, medium, high)
```

```python
# SIMULATION
results = []

for e in elasticities:
    for ubi in ubi_levels:
        delta_income = ubi
        percent_change_income = delta_income / base_income
        percent_change_hours = -e * percent_change_income  # negative because income ↑ → hours ↓

        results.append({
            'Elasticity': e,
            'UBI ($)': ubi,
            'Labor Supply Change (%)': percent_change_hours * 100  # in %
        })

# Create DataFrame
df = pd.DataFrame(results)
```

```python
# DISPLAY RESULTS
print(df.to_string(index=False))
```

```python
# PLOT THE RESULTS
plt.figure(figsize=(10, 6))

for e in elasticities:
    subset = df[df['Elasticity'] == e]
    plt.plot(
        subset['UBI ($)'],
        subset['Labor Supply Change (%)'],
        marker='o',
        label=f'Elasticity = {e}'
    )

plt.title('Labor Supply Change vs. UBI Levels (Sensitivity Analysis)', fontsize=14)
plt.xlabel('UBI Level ($ per year)', fontsize=12)
plt.ylabel('Labor Supply Change (%)', fontsize=12)
plt.grid(True)
plt.legend(title='Elasticity')
plt.tight_layout()
plt.show()
```

Interpretation of Simulation Results
With a labor-supply elasticity of 0.10, a $12,000 UBI reduces total work hours by about 2.4%.

With higher elasticities, the drop in labor supply becomes larger:

At ε = 0.05, UBI cuts labor supply only ~1.2%.

At ε = 0.20, the same UBI cuts labor supply by ~4.8%.

In simple terms: the more sensitive workers are to extra income, the more they cut hours. That means UBI has a stronger effect if people react more to free income.
